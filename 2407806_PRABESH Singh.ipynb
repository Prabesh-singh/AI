```python
# Task - 3 - Outlier Detection:
# 1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
```

```python
score_q1 = south_asia_df['score'].quantile(0.25)
score_q3 = south_asia_df['score'].quantile(0.75)
score_iqr = score_q3 - score_q1

gdp_q1 = south_asia_df['Log GDP per capita'].quantile(0.25)
gdp_q3 = south_asia_df['Log GDP per capita'].quantile(0.75)
gdp_iqr = gdp_q3 - gdp_q1

# Step 2: Define outlier bounds
score_lower_bound = score_q1 - 1.5 * score_iqr
score_upper_bound = score_q3 + 1.5 * score_iqr

gdp_lower_bound = gdp_q1 - 1.5 * gdp_iqr
gdp_upper_bound = gdp_q3 + 1.5 * gdp_iqr

# Step 3: Identify outliers based on these bounds
score_outliers = south_asia_df[(south_asia_df['score'] < score_lower_bound) | (south_asia_df['score'] > score_upper_bound)]
gdp_outliers = south_asia_df[(south_asia_df['Log GDP per capita'] < gdp_lower_bound) | (south_asia_df['Log GDP per capita'] > gdp_upper_bound)]

# Step 4: Display the outliers
print("Outliers based on Score:")
print(score_outliers)

print("\nOutliers based on GDP per Capita:")
print(gdp_outliers)
```

```python
# Calculate IQR for Happiness Score
score_q1 = south_asia_df['score'].quantile(0.25)
score_q3 = south_asia_df['score'].quantile(0.75)
score_iqr = score_q3 - score_q1

# Calculate IQR for GDP per Capita
gdp_q1 = south_asia_df['Log GDP per capita'].quantile(0.25)
gdp_q3 = south_asia_df['Log GDP per capita'].quantile(0.75)
gdp_iqr = gdp_q3 - gdp_q1

# Step 2: Define outlier bounds
score_lower_bound = score_q1 - 1.5 * score_iqr
score_upper_bound = score_q3 + 1.5 * score_iqr

gdp_lower_bound = gdp_q1 - 1.5 * gdp_iqr
gdp_upper_bound = gdp_q3 + 1.5 * gdp_iqr

# Step 3: Identify outliers based on these bounds
outliers = south_asia_df[
    (south_asia_df['score'] < score_lower_bound) | 
    (south_asia_df['score'] > score_upper_bound) |
    (south_asia_df['Log GDP per capita'] < gdp_lower_bound) |
    (south_asia_df['Log GDP per capita'] > gdp_upper_bound)
]

# Step 4: Create the scatter plot
plt.figure(figsize=(10, 6))

# Plotting GDP vs. Score
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], color='blue', label='Normal Data')

# Highlighting the outliers
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label='Outliers' )
for i, row in south_asia_df.iterrows(): # Add label for each countries
    plt.text(row['Log GDP per capita'] + 0.02, row['score'], row['Country name'], fontsize=9)
# Adding labels and title
plt.xlabel('Log GDP per capita')
plt.ylabel('Score')
plt.title('Log GDP per capita vs. Score (Outliers Highlighted)')

# Show the legend
plt.legend()

# Display the plot
plt.show()

# Step 5: Display the outliers
print("Outliers based on GDP per Capita or Score:")
print(outliers)
```

```python
# Calculate Q1 and Q3 for Score and Log GDP per Capita
Q1_score = middle_east_df['score'].quantile(0.25)
Q3_score = middle_east_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = middle_east_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = middle_east_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Find lower and upper outlier
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers
middle_east_score_outliers = middle_east_df[(middle_east_df['score'] < lower_bound_score) | (middle_east_df['score'] > upper_bound_score)]
middle_east_gdp_outliers= middle_east_df[(middle_east_df['Log GDP per capita'] < lower_bound_gdp) | (middle_east_df['Log GDP per capita'] > upper_bound_gdp)]

# Check if there are any outliers
if not middle_east_score_outliers.empty:
    print(f"As per score outlier countries in the Middle East:\n {middle_east_score_outliers[['Country name', 'score']]}")
else:
    print("There are no score outliers in the Middle East.")

if not middle_east_gdp_outliers.empty:
    print(f"As per GDP outlier countries in the Middle East:\n {middle_east_gdp_outliers[['Country name', 'Log GDP per capita']]}")
else:
    print("There are no GDP outliers in the Middle East.")
```
